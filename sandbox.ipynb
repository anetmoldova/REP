```python
import requests
from bs4 import BeautifulSoup
```

```python
url = f"https://www.bezrealitky.cz/vypis/nabidka-pronajem/byt/praha?page"
response = requests.get(url)

response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
cards = soup.find_all('article', class_='listing-card')

for card in cards:
    title = card.find('h2', class_='listing-title')
    price = card.find('div', class_='price')
    location = card.find('div', class_='listing-location')
    
    print("NÃ¡zov:", title.text.strip() if title else "N/A")
    print("Cena:", price.text.strip() if price else "N/A")
    print("Lokalita:", location.text.strip() if location else "N/A")
    print("---")
```
