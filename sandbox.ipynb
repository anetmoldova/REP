```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = f"https://www.bezrealitky.cz/vypis/nabidka-pronajem/byt/praha?page"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# 'title' contains all titles of the listings and the link to the listing
title= soup.find_all('h2', class_='PropertyCard_propertyCardHeadline___diKI mt-md-0 mt-4 mb-0')

# 'price' contains prices of the listings
price_list = soup.find_all('div', class_="PropertyPrice_propertyPrice__lthza propertyPrice mb-0 mt-3")

rent = [price_list[i].find('span', class_="PropertyPrice_propertyPriceAmount__WdEE1").text.replace('\xa0', '') for i in range(len(price_list))]

#href contains all url to the listings
hrefs_listings_list = [title[i].find('a')['href'] for i in range(len(title))]
```

```python
# Merge lists into a dictionary to make a dataframe
listings_info_dict = {
    'url': hrefs_listings_list,
    'price': rent,
}

df = pd.DataFrame.from_dict(listings_info_dict)
```

```python
import requests

def geocode_osm(address):
    params = {"q": address, "format": "json", "limit": 1}
    r = requests.get("https://nominatim.openstreetmap.org/search", params=params)
    r.raise_for_status()
    data = r.json()
    if data:
        lat = data[0]["lat"]
        lon = data[0]["lon"]
        return lat, lon
    return None

address = "Praha Újezd nad Lesy Pod Klaudiánkou 861057"
coords = geocode_osm(address)
print(coords)
```

```python
poi_items = soup.find_all('div', class_='Poi_poiItem__7JgIz')

for item in poi_items:
    try:
        typ_tag = item.find('span', class_='Poi_poiItemContentType__XukbX poiItemContentType')
        typ = typ_tag.get_text(strip=True) if typ_tag else '???'

        nazov_tag = item.find('strong', class_='poiItemContentName')
        nazov = nazov_tag.get_text(strip=True) if nazov_tag else '???'

        time_block = item.find('div', class_='Poi_poiItemTimes__hse64')
        vzdialenost_tag = time_block.find('strong') if time_block else None
        vzdialenost = vzdialenost_tag.get_text(strip=True) if vzdialenost_tag else '???'

        cas_text = time_block.get_text(strip=True) if time_block else ''
        cas = cas_text.split('(')[-1].replace(')', '') if '(' in cas_text else '???'

        print(f"{typ}: {nazov} – {vzdialenost}, {cas}")
    
    except Exception as e:
        print(f"Chyba pri spracovaní: {e}")
```
